## Treating the Outliears -

In this plots we can see the their are many Outliers in the variables , so we first see and then later on will treat the outliers with the IQR method.

```python
plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x = 'BHK', y = 'Rent', data = data)
plt.subplot(2,3,2)
sns.boxplot(x = 'Bathroom', y = 'Rent', data = data)
plt.subplot(2,3,3)
sns.boxplot(x = 'Size', y = 'Rent', data = data)

plt.show()
```

```python
# Here we are treating the outliers with the IQR method with the all 4 variables who has outliers-

cont_numeric=["Rent","Size"]
desc_numeric=["BHK","Bathroom"]

for i in cont_numeric:
    Q1=np.percentile(data[i],25,interpolation="midpoint")
    Q3=np.percentile(data[i],75,interpolation="midpoint")
    IQR=Q3-Q1
    up_lim=Q3+(1.5* IQR)
    low_lim=Q1-(1.5* IQR)
    
    data.loc[data[i]>up_lim,i]=up_lim

data[cont_numeric].boxplot()
```

# Step 3: Data Preparation-

 You can see that your dataset has many columns with categorical values(options).
But in order to fit a regression line, we would need numerical values and not string. Hence, we need to convert them to numbers in order to their options.

Here we are doing the Label encoding , removing unwanted parts, converting floor columns into total floor etc

```python
#Droping Posted on Column

data.drop("Posted On", axis=1, inplace=True)
```

```python
# We split the text data in the Floor column and take the first value(eg. Ground out of 2/ 1 out of 30 etc)

# eg. 1 out of 10 : here value 1(rent floor) will be stored in the 10th floor.

Rent_Floor=data["Floor"].str.split(pat=" ",n=1,expand=True).iloc[:,0]
Rent_Floor.value_counts()
```

```python
# Our data contains Ground , Upper and Lower which is not a numeric 
# so we replace that Ground = 0.5 , Upper = 0.3 and Lower = 0.1

Rent_Floor=Rent_Floor.replace("Ground",'0.5')
Rent_Floor=Rent_Floor.replace("Upper",'0.3')
Rent_Floor=Rent_Floor.replace("Lower",'0.1')

Rent_Floor
#
```

```python
# Total number of floor available in apartment-  
# eg. 1 out of 1o : so in this step we will store the value 10 in the total no of floor)

Total_flr=data["Floor"].str.split(pat=" ",n=3,expand=True).iloc[:,3]

Total_flr.value_counts()
```

```python
# It contain 'of' so we will remove them

Total_floor=[]

for i,j in enumerate(Total_flr):
    k=str(j).strip('of')
    Total_floor.append(k)
```

```python
# Here we have added the Rent_floor and Total_floor in out dataset 

data["Rent_Floor"]=pd.DataFrame(Rent_Floor)
data['Total_Floor'] = pd.DataFrame(Total_floor)
```

```python
# Here we have drop the Area Locality and Floor column 

data.drop(["Area Locality","Floor"],axis=1,inplace=True)
data
```

Checking VIF
Variance Inflation Factor or VIF, gives a basic quantitative idea about how much the feature variables are correlated with each other. It is an extremely important parameter to test our linear model. The formula for calculating VIF is:

ð‘‰ð¼ð¹ð‘–=11âˆ’ð‘…ð‘–2

```python
# case 1: low p-value and low VIF -- keep

# case 2 : High p-value and high VIF -- remove

# case 3 : a) High p-value and low VIF -- remove first
        #  b) Low p-value and high VIF -- remove later
```

```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#Dropping the variable and updating the model
#As you can see from the summary and the VIF dataframe, some variables are still insignificant. One of these variables is, semi-furnished as it has a very high p-value of 0.938. 
#Let's go ahead and drop this variables
```
